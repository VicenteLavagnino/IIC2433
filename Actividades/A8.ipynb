```python
import pandas as pd
import numpy as np
import kagglehub


# download latest version
path = kagglehub.dataset_download("mashlyn/online-retail-ii-uci")

foo = path + '/online_retail_II.csv'

print(foo)

df = pd.read_csv(foo, encoding="latin1")

data = df.groupby("Invoice")["StockCode"].apply(list).tolist()
data = data[:50_000]
```

```python
from mlxtend.preprocessing import TransactionEncoder

a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df_apriori = pd.DataFrame(a_data, columns=a.columns_)
df_apriori = df_apriori.replace(False, 0)
df_apriori = df_apriori.replace(True, 1)
df_apriori
```

```python
df = df_apriori
```

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules

df_apriori = apriori(df, min_support=0.2, use_colnames=True, verbose=1)
df_apriori
```
